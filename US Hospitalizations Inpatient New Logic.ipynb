```python
import pandas as pd
import numpy as np
import datetime as dt
import calendar
from dateutil.relativedelta import relativedelta



HHS_NHSN = pd.read_excel(r'C:\Users\ssing296\Desktop\Python-Coursera\HHS NHSN COVID Data1.xlsx')
WOM_US_Rates = pd.read_excel(r'C:\Users\ssing296\Desktop\Python-Coursera\WOM US Rates Data5.xlsx')

#WOM_US_Rates

def hospital_bed(df,per_val,col_header):
    
    
    per_value = float(per_val)
    

    
    hosp_states = df['State'].iloc[0:50].tolist()
    hosp_values = df[col_header].iloc[0:50].tolist()

    hosp_dict = {}
    hosp_other_dict = {}
    
#   hosp_states = df[round(df[col_header],0) >= per_value]['State'].iloc[0:50].tolist()
#   hosp_values = df[round(df[col_header],0) >= per_value][col_header].iloc[0:50].tolist()


     #dict(sorted(test_dict.items(), key = itemgetter(1), reverse = True)[:N]) 
    

    for i in range(0,len(hosp_states)):
        if(round(hosp_values[i]*100,0) >= per_value):
            hosp_dict[hosp_states[i]] = hosp_values[i]
        else:    
            hosp_other_dict[hosp_states[i]] = hosp_values[i]
         
    

    hosp_dict = {k: v for k, v in sorted(hosp_dict.items(),reverse = True, key=lambda x: x[1])} 
    hosp_other_dict = {k: v for k, v in sorted(hosp_other_dict.items(),reverse = True, key=lambda x: x[1])} 
    
    #print(hosp_dict)
    
    #print("-------------------------------------------------------")
    
    #print(hosp_other_dict)
    
    print_text = ''
    
    if(col_header == 'Percentage of Inpatient Beds Occupied Estimated'):
        print_text = 'Inpatient'
    else:
        print_text = 'ICU'
        
    if len(hosp_dict) == 0:
        a = ''
    
    elif len(hosp_dict) == 1:
        state1 = list(hosp_dict)[0]
        state1_val = round(list(hosp_dict.values())[0]*100)
      
        a = 'One State had an {} bed occupancy greater than {:}%: {} ({:}%)'.format(print_text,per_value,state1,state1_val)
        
    
    elif len(hosp_dict) == 2:
        state1 = list(hosp_dict)[0]
        state1_val = round(list(hosp_dict.values())[0]*100)

        state2 = list(hosp_dict)[1]
        state2_val = round(list(hosp_dict.values())[1]*100)
        
        a = 'Two States had an {} bed occupancy greater than {:}%: {} ({:}%) and {} ({:}%)'.format(print_text,per_value,state1,state1_val,state2,state2_val)
        
    elif len(hosp_dict) == 3:
        state1 = list(hosp_dict)[0]
        state1_val = round(list(hosp_dict.values())[0]*100)

        state2 = list(hosp_dict)[1]
        state2_val = round(list(hosp_dict.values())[1]*100)

        state3 = list(hosp_dict)[2]
        state3_val = round(list(hosp_dict.values())[2]*100)

        
        a = 'Three States had an {} bed occupancy greater than {:}%: {} ({:}%), {} ({:}%) and {} ({:}%)'.format(print_text,per_value,state1,state1_val,state2,state2_val,state3,state3_val)
        

    elif len(hosp_dict) == 4:
        state1 = list(hosp_dict)[0]
        state1_val = round(list(hosp_dict.values())[0]*100)

        state2 = list(hosp_dict)[1]
        state2_val = round(list(hosp_dict.values())[1]*100)

        state3 = list(hosp_dict)[2]
        state3_val = round(list(hosp_dict.values())[2]*100)

        state4 = list(hosp_dict)[3]
        state4_val = round(list(hosp_dict.values())[3]*100)

        
        a = 'Four States had an {} bed occupancy greater than {:}%: {} ({:}%), {} ({:}%), {} ({:}%) and {} ({:}%)'.format(print_text,per_value,state1,state1_val,state2,state2_val,state3,state3_val,state4,state4_val)

    elif len(hosp_dict) >= 5:
        state1 = list(hosp_dict)[0]
        state1_val = round(list(hosp_dict.values())[0]*100)

        state2 = list(hosp_dict)[1]
        state2_val = round(list(hosp_dict.values())[1]*100)

        state3 = list(hosp_dict)[2]
        state3_val = round(list(hosp_dict.values())[2]*100)

        state4 = list(hosp_dict)[3]
        state4_val = round(list(hosp_dict.values())[3]*100)
        
        state5 = list(hosp_dict)[4]
        state5_val = round(list(hosp_dict.values())[4]*100)

        
        a = 'Five States had an {} bed occupancy greater than {:}%: {} ({:}%), {} ({:}%), {} ({:}%),{} ({:}%) and {} ({:}%)'.format(print_text,per_value,state1,state1_val,state2,state2_val,state3,state3_val,state4,state4_val,state5,state5_val)
    
    
    if len(hosp_dict) == 0:
        hosp_other_states1 = list(hosp_other_dict)[0]
        hosp_other_states1_val = round(list(hosp_other_dict.values())[0]*100)

        hosp_other_states2 = list(hosp_other_dict)[1]
        hosp_other_states2_val = round(list(hosp_other_dict.values())[1]*100)

        hosp_other_states3 = list(hosp_other_dict)[2]
        hosp_other_states3_val = round(list(hosp_other_dict.values())[2]*100)

        hosp_other_states4 = list(hosp_other_dict)[3]
        hosp_other_states4_val = round(list(hosp_other_dict.values())[3]*100)
        
        hosp_other_states5 = list(hosp_other_dict)[4]
        hosp_other_states5_val = round(list(hosp_other_dict.values())[4]*100)
        
        b = 'Five States had a rate of new cases per 100,000 population: {} ({:}%), {} ({:}%), {} ({:}%),{} ({:}%) and {} ({:}%).'.format(hosp_other_states1,hosp_other_states1_val,hosp_other_states2,hosp_other_states2_val,hosp_other_states3,hosp_other_states3_val,hosp_other_states4,hosp_other_states4_val,hosp_other_states5,hosp_other_states5_val)
    
    elif len(hosp_dict) == 1:
        hosp_other_states1 = list(hosp_other_dict)[0]
        hosp_other_states1_val = round(list(hosp_other_dict.values())[0]*100)

        hosp_other_states2 = list(hosp_other_dict)[1]
        hosp_other_states2_val = round(list(hosp_other_dict.values())[1]*100)

        hosp_other_states3 = list(hosp_other_dict)[2]
        hosp_other_states3_val = round(list(hosp_other_dict.values())[2]*100)

        hosp_other_states4 = list(hosp_other_dict)[3]
        hosp_other_states4_val = round(list(hosp_other_dict.values())[3]*100)
        b = ' with the next four highest States being {} ({:}%), {} ({:}%),{} ({:}%) and {} ({:}%).'.format(hosp_other_states1,hosp_other_states1_val,hosp_other_states2,hosp_other_states2_val,hosp_other_states3,hosp_other_states3_val,hosp_other_states4,hosp_other_states4_val)
        
    elif len(hosp_dict) == 2:
        hosp_other_states1 = list(hosp_other_dict)[0]
        hosp_other_states1_val = round(list(hosp_other_dict.values())[0]*100)

        hosp_other_states2 = list(hosp_other_dict)[1]
        hosp_other_states2_val = round(list(hosp_other_dict.values())[1]*100)

        hosp_other_states3 = list(hosp_other_dict)[2]
        hosp_other_states3_val = round(list(hosp_other_dict.values())[2]*100)

        b = ' with the next three highest States being {} ({:}%),{} ({:}%) and {} ({:}%).'.format(hosp_other_states1,hosp_other_states1_val,hosp_other_states2,hosp_other_states2_val,hosp_other_states3,hosp_other_states3_val)

    elif len(hosp_dict) == 3:
        hosp_other_states1 = list(hosp_other_dict)[0]
        hosp_other_states1_val = round(list(hosp_other_dict.values())[0]*100)

        hosp_other_states2 = list(hosp_other_dict)[1]
        hosp_other_states2_val = round(list(hosp_other_dict.values())[1]*100)

        b = ' with the next two highest States being {} ({:}%) and {} ({:}%).'.format(hosp_other_states1,hosp_other_states1_val,hosp_other_states2,hosp_other_states2_val)

    elif len(hosp_dict) == 4:
        hosp_other_states1 = list(hosp_other_dict)[0]
        hosp_other_states1_val = round(list(hosp_other_dict.values())[0]*100)


        b = ' with the next highest State being {} ({:}%).'.format(hosp_other_states1,hosp_other_states1_val)

    else:
        b = ''
        

    return a+b




        
States_Comments_Hospital_Inpatient_Bed = hospital_bed(HHS_NHSN,85,'Percentage of Inpatient Beds Occupied Estimated')
States_Comments_Hospital_ICU_Bed = hospital_bed(HHS_NHSN,90,'Percentage of ICU Beds Occupied Estimated')



        
print(States_Comments_Hospital_Inpatient_Bed)
print(States_Comments_Hospital_ICU_Bed)
```
