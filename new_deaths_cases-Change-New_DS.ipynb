```python
import pandas as pd
import numpy as np
import datetime as dt
import calendar
from dateutil.relativedelta import relativedelta



df = pd.read_excel(r'C:\Users\ssing296\Desktop\Python-Coursera\COVID_QA_Final Output.xlsx')
global_cases = pd.read_excel(r'C:\Users\ssing296\Desktop\Python-Coursera\OCT_Worldometer Worldwide New Cases_OD_SafetoShare.xlsx')
global_deaths = pd.read_excel(r'C:\Users\ssing296\Desktop\Python-Coursera\OCT_Worldometer Worldwide New Deaths_OD_SafetoShare.xlsx')

usa_cases = pd.read_excel(r'C:\Users\ssing296\Desktop\Python-Coursera\US Based Worldometer New Cases.xlsx')
usa_deaths = pd.read_excel(r'C:\Users\ssing296\Desktop\Python-Coursera\WOM_COVID_US_NewDeath_MovingAverage.xlsx')


def new_cases_KPI1 (df,section,group):
    
  
    if(section == 'U.S. Deaths'):
        col_selection = 'Overall New Deaths'
    elif(section == 'U.S. Cases'):
        col_selection = 'United States'
    else:
        col_selection = 'Worldwide'

    if(section == 'U.S. Deaths'):
        datecol = 'Date Updated'
    else:
        datecol = 'Date'


    maxdate = max(df[datecol])
    
    week_ago = maxdate - dt.timedelta(days=7)
    
    four_week_ago = maxdate - dt.timedelta(days=28)

    current_worldwide = int(df[(df[datecol] == maxdate)][col_selection].sum())
    weekago_worldwide = int(df[(df[datecol] == week_ago)][col_selection].sum())
    
    per_change = (current_worldwide - weekago_worldwide) / weekago_worldwide
    
    if(per_change >= 0):
        print_text = "up"
    else:
        print_text = "down"
        
    a ='{:,} new {} on {}/{}, {} {:.1f}% from {:,} last {}.'.format(current_worldwide,group,maxdate.month,maxdate.day,print_text,abs(per_change)*100,weekago_worldwide,calendar.day_name[week_ago.weekday()])

    
    if((section == 'global') & (group == 'deaths')):
        macol = 'New Deaths N Days Moving AVG'
    else:
        macol = 'New Cases N Days Moving AVG'

    
    current_ma = int(df[(df[datecol] == maxdate)][macol].sum())
    weekago_ma = int(df[(df[datecol] == week_ago)][macol].sum())
    monthage_ma = int(df[(df[datecol] == four_week_ago)][macol].sum())
    
    change_per_wow = (current_ma-weekago_ma) / weekago_ma
    change_per_mom = (current_ma-monthage_ma) / monthage_ma
    
    if change_per_wow >= 0:
        wow_text = "up"
    else:
        wow_text = "down"
    
    if change_per_mom >= 0:
        mom_text = "up"
    else:
        mom_text = "down"

    b = '{:,.0f} ({} {:.1f}% WoW) nominal 7-day moving average of new {} on {}/{}, {} {:.1f}% from {:,.0f} one month ago.'.format(current_ma,wow_text,abs(change_per_wow)*100,group,maxdate.month,maxdate.day,mom_text,abs(change_per_mom)*100,monthage_ma)
    
    
    #format(MA_yday,print_text_ma_pc,abs(MA_pc)) + group+
    
    return a,b

Global_Comment_Cases1 = new_cases_KPI1(global_cases,'global','cases')
Global_Comment_Deaths1 = new_cases_KPI1(global_deaths,'global','deaths')

US_Comment_Cases1 = new_cases_KPI1(usa_cases,'U.S. Cases','cases')
US_Comment_Deaths1 = new_cases_KPI1(usa_deaths,'U.S. Deaths','deaths')



print("******************************************USA*********************************************")
print("US New Cases -> "+ US_Comment_Cases1[0])
print("US New Cases -> "+ US_Comment_Cases1[1])

print("US New Deaths -> "+ US_Comment_Deaths1[0])
print("US New Deaths -> "+ US_Comment_Deaths1[1])
print("******************************************Global*********************************************")

print("Global New Cases -> "+ Global_Comment_Cases1[0])

print("Global New Cases -> "+ Global_Comment_Cases1[1])


print("Global New Deaths -> "+ Global_Comment_Deaths1[0])

print("Global New Deaths -> "+ Global_Comment_Deaths1[1])
```
